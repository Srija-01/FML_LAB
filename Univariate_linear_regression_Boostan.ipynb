```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv("boston.csv")
data.head()
```

```python
X = data.drop('medv', axis=1).values
y = data['medv'].values
```

```python
X_rooms = X[:, 5]
print(X_rooms)
```

```python
type(X_rooms), type(y)
```

```python
X_rooms.shape, y.shape
```

```python
X_rooms = X_rooms.reshape(-1,1)
y = y.reshape(-1,1)
print(y)
```

```python
X_rooms.shape, y.shape
```

```python
plt.scatter(X_rooms, y)
plt.xlabel('Number of rooms')
plt.ylabel('Value of house ?1000($)')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_rooms, y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred = reg.predict(X_rooms)
print(f'MAE = {mean_absolute_error(y, y_pred)}')
print(f'MSE = {mean_squared_error(y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y, y_pred))}')
print(f'R_2 = {r2_score(y, y_pred)}')
```

```python
prediction_space = np.linspace(min(X_rooms), max(X_rooms)).reshape(-1,1)
print(prediction_space)
```

```python
plt.scatter(X_rooms, y, color="black")
plt.plot(X_rooms, y_pred, color="red", linewidth=3)
plt.show()
```

```python
plt.scatter(X_rooms, y, color="black")
plt.plot(prediction_space, reg.predict(prediction_space), color="yellow", linewidth=3)
plt.show()
```
